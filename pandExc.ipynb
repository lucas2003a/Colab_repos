```python
#EJERCICIO 1
import pandas as pd
finicio = int(input("Introduce el año incial: "))
ffin = int(input("Introduce el año final. "))

ventas = {}

for i in range(finicio, ffin+1):
  ventas[i] = float(input("Introduce las ventas de la año: " + str(i) + ": "))

ventas = pd.Series(ventas)

print("ventas\n", ventas)
print("ventas con descuento\n", ventas*0.9)
```

```python
#EJERCICIO 2

## solución 1
def estadistica_notas(notas):
  notas = pd.Series(notas)
  estadisticas = pd.Series([notas.min(),notas.max(),notas.mean(),notas.std()], index = ["Máximo","Mínimo","Promedio","Desviación"])
  return estadisticas

notas = {"alonso": 19,"Yorghet": 15,"Adriana": 20,"Lucas": 18}
print(estadistica_notas(notas))
```

```python
##solución 2

def estadistica_notasV2(notas):
  notas = pd.Series(notas)
  return notas.describe()

notas = {"alonso": 19,"Yorghet": 15,"Adriana": 20,"Lucas": 18}
print(estadistica_notasV2(notas))
```

```python
##ejercicio 3
def aprobados(notas):
  notas = pd.Series(notas)
  return notas[notas >=11].sort_values(ascending = False)

notas = {"alonso": 19,"Yorghet": 15,"Adriana": 20,"Lucas": 18,"Juan": 13,"Manuel":9,"Tito":6}
print(aprobados(notas))
```

```python
#ejercicio 5
##solucion 1

datos = {"Mes":["Enero","Febreo","Marzo","Abril","Mayo","Junio"], "Ventas":[30050,62245,45963,4568512,45612,15632],"Gastos":[12320,5213,312,456,456,876]}
contabilidad = pd.DataFrame(datos)

def balance(contabilidad, meses):
  contabilidad["Balance"] = contabilidad.Ventas = contabilidad.Gastos
  return contabilidad[contabilidad.Mes.isin(meses)].Balance.sum()

print(balance(contabilidad,["Enero","Marzo"]))
```

```python
##solucion 2
  ##Usa datos y contabilidad del ejemplo anterior

def balance2(contabilidad, meses):
  contabilidad["Balance"] = contabilidad.Ventas = contabilidad.Gastos
  return contabilidad.set_index("Mes").loc[meses,"Balance"].sum()

print(balance2(contabilidad,["Enero","Marzo"]))
```

```python
#ejercicio 6
def resume_cotizaciones(fichero):
  df = pd.read_csv(fichero, sep = ";", decimal = ",", thousands =  ".",index_col = 0)
  return pd.DataFrame([df.min(),df.max(),df.mean()], index = ["Mínimo","Máximo","Media"])

resume_cotizaciones("cotizacion.csv")
```
