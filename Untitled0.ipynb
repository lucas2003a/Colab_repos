```python
import pandas as pd

datos = pd.read_csv("creditos.csv", sep = ";")
```

```python
import matplotlib.pyplot as mat

mat.ylabel("EtiquetaY")
mat.xlabel("EtiquetaX")

mat.scatter(datos["credito"], datos["cumplio"], color = "green")
mat.show()
```

```python
from sklearn import linear_model

regresion = linear_model.LinearRegression()

credito = datos["credito"].values.reshape((-1,1))

modelo = regresion.fit(credito,datos["cumplio"])
```

```python
entrada = [[0],[1]]

prediccion = modelo.predict(entrada)

mat.scatter(entrada,prediccion, color = "blue")
mat.plot(entrada,prediccion, color="black")

mat.xlabel("credito")
mat.ylabel("cumpliÃ³")

mat.scatter(datos["credito"], datos["cumplio"], color= "green")
mat.show()

```
