```python
#importando la librería de lectura de csv
import pandas as panda

datos = panda.read_csv("carros_csv.csv", sep = ";")
#datos
```

```python
#Generar la base del gráfico
import matplotlib.pyplot as mat

#Configuración de los ejes x-y
mat.ylabel("Etiqueta Y")
mat.xlabel("Etiqueta X")

#Enviar datos al gráfico
mat.scatter(datos["kms"], datos["precio"], color = "red")
mat.show()
```

```python
#Calcular la intersección y pendiente | modelo de análisis de regresión lineal
from sklearn import linear_model

#Objeto que contien el modelo RL(Regresión Lineal)
regresion = linear_model.LinearRegression();

#Variable independiente (X). kms
kms = datos["kms"].values.reshape((-1,1)) #reformar/acondicionar/preparar

#Variable dependiente (Y). precio

#Entrenamiento (..Vamos en busqueda de Y)
modelo  = regresion.fit(kms, datos["precio"])


#Tenemos los datos()
print("Intersección", modelo.intercept_)
print("Pendiente", modelo.coef_)
```

```python
entrada = [[6048],[10000],[42000],[5200],[139000]]
prediccion = modelo.predict(entrada)

#Renderizamos estos resultados en un gráfico
mat.scatter(entrada, prediccion, color ="blue") # Render 1 = ecuación estimada
mat.plot(entrada, prediccion, color ="black") #Render 2 = línea
mat.ylabel("Título Y")
mat.xlabel("Título X")
mat.scatter(datos["kms"],datos["precio"],color="green") # Render
mat.show()
```
