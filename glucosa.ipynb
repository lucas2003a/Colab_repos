```python
import pandas as pd
```

```python
datos = pd.read_csv("diabetes.csv", sep = ",")
```

```python
import matplotlib.pyplot as mat

mat.xlabel("EtiquetaX")
mat.ylabel("EtiquetaY")

mat.scatter(datos["imc"], datos["glucosa"], color = "red")
mat.show()
```

```python
#Calculos

from sklearn import linear_model

regresion = linear_model.LinearRegression()

imc = datos["imc"].values.reshape((-1,1))

modelo = regresion.fit(imc, datos["glucosa"])

print("intersección",modelo.intercept_)
print("Pendiente", modelo.coef_)
```

```python
entrada = [[10],[15],[20],[25],[30]]

prediccion = modelo.predict(entrada)

mat.scatter(entrada,prediccion,color = "blue")
mat.plot(entrada,prediccion, color = "black")
mat.ylabel("Título X")
mat.xlabel("Título Y")

mat.scatter(datos["imc"],datos["glucosa"], color="green")
mat.show();
```
